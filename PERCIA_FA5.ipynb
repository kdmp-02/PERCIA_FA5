```python
import time

def recursion_countdown(n):
    if n <= 0:
        print('Blastoff!')
    else:
        print(n)
        recursion_countdown(n-1)

def while_countdown(n):
    while n > 0:
        print(n)
        n = n - 1
    print('Blastoff!')
```

```python
#For n=100

a = 100

rstart100 = time.time()
recursion_countdown(a)
rend100 = time.time()
rtime100 = float(rend100 - rstart100)

istart100 = time.time()
while_countdown(a)
iend100 = time.time()
itime100 = float(iend100 - istart100)

print("- - - - - - - - - - - - - ")
    
print("\nRecursion Time : " + str(rtime100))
print("\nWhile Time : " + str(itime100))
```

```python
#For n=1000

b = 1000

rstart1000 = time.time()
recursion_countdown(b)
rend1000 = time.time()
rtime1000 = float(rend1000 - rstart1000)

istart1000 = time.time()
while_countdown(b)
iend1000 = time.time()
itime1000 = float(iend1000 - istart1000)

print("- - - - - - - - - - - - - ")
    
print("\nRecursion Time : " + str(rtime1000))
print("\nWhile Time : " + str(itime1000))
```

```python
#For n=10000

c = 10000

rstart10000 = time.time()
recursion_countdown(c)
rend10000 = time.time()
rtime10000 = float(rend10000 - rstart10000)

istart10000 = time.time()
while_countdown(c)
iend10000 = time.time()
itime10000 = float(iend10000 - istart10000)

print("- - - - - - - - - - - - - ")
    
print("\nRecursion Time : " + str(rtime10000))
print("\nWhile Time : " + str(itime10000))
```

From these results, we can see that the while loop implementation is significantly faster than the recursive implementation for all values of n. This is because the while loop implementation avoids the function call overhead and stack usage that comes with recursion.
